```python
from scipy import stats as st
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
from functools import partial
from statsmodels.stats.diagnostic import lilliefors
```

## Задача 1.1
Проверить мощность и консервативность критериев Лиллиефорса, Харке-Бера, Шапиро-Улика для выборок из следующих распределений:
* Нормальное
* Лапласа
* Стьюдента
* Усеченное нормальное распределение (модуль каждого элемента выборки не превосходит 2)


```python
n_of_h = 10000
alpha = 0.05
```

```python
dists = [(st.norm.rvs, 'normal'), 
         (st.laplace.rvs, 'laplace'), 
         (st.t.rvs, 'student'), 
         (partial(st.truncnorm.rvs, a=-2, b=2), 'truncated normal')]
criterions = [lilliefors, st.jarque_bera, st.shapiro]
```

```python
for criterion in criterions:
    print(f'Criterion: {criterion.__name__}')
    false_negative_ps = []
    for dist_gen, dist_name in dists:
        print(f'\tDistribution: {dist_name}')
        for volume in [10, 20, 100, 500, 1000, 2000, 5000]:
            print(f'\t\tSize {volume},', end=' ')
            # power
            in_critical = 0
            for j in range(n_of_h):
                if dist_name == 'student':
                    k = np.random.randint(1, 50)
                    sample = dist_gen(size=volume, df=k)
                else:
                    loc = (np.random.rand() - 0.5) * 100
                    scale = np.random.rand() * 100
                    sample = dist_gen(loc=loc, scale=scale, size=volume)
                if criterion(sample)[1] <= alpha:
                    in_critical += 1
            prob_in_critical = in_critical / n_of_h
            print(f'\tprobability of critical area = {prob_in_critical}', end=' ')
            if dist_name == 'normal':
                print(f' -- false negative (conservative if < {alpha})')
            else:
                print(' -- power')
```
