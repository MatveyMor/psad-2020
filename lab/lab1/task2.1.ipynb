```python
from numpy import random as rand
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline

import scipy.stats as st
import pandas as pd
import seaborn as sns

from tqdm import tqdm
```

```python
p_variance = np.linspace(0, 1, 88)
p_0 = 0.25
alpha = 0.05
p_values = []
power_of_tests = []
list_of_sizes = list(map(int, np.linspace(10, 1000, 100)))
for size in tqdm(list_of_sizes):
    w = 0
    for p in p_variance:
        p_value = st.binom_test(size * p, n = size, p = p_0, alternative='two-sided')
        p_values.append(p_value)
        if p_value < alpha:
            w += 1
    power_of_tests.append(w/len(p_variance))
```

```python
plt.plot(list_of_sizes, power_of_tests)
plt.xlabel('Size of set')
plt.ylabel('Power')
plt.savefig('p(n).png')
```
