# №1

Ошибка первого рода в данном случае интерпетируется как ложное срабатывание, ошибка второго рода - пропуск подозреваемого.
Отвегая нулевые гипотезы на одном уровне $alpha = 0.05$ и не учитывая множественность гипотез, мы не можем быть уверены, что $FWER <= \alpha$, а в данном случае $FWER$ есть вероятность ошибки первого рода для классификатора в целом. Обеспечивать контроль над $FWER$ позволяют такие способы, как например метод Холма.
